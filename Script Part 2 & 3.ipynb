```python
### keuntungan yang didapat dari OK! Food Delivery setiap harinya
query3 = """

SELECT 
	    order_date,
	    SUM(price*quantity) AS Total_Keuntungan
From  `aqueous-freedom-398208.dkf_okefood.order` O
JOIN `aqueous-freedom-398208.dkf_okefood.food` F ON O.fk_food_id = F.food_id
GROUP BY order_date
ORDER BY order_date ASC


  """

#Run the query and write result to a pandas data frame

result = bigquery_client.query(query3)

df = result.to_dataframe()

#View top few rows of result

df.head(15)
```

```python
df.info()
```

```python
### average spending 1 orang customer setiap harinya dalam 1 bulan
query4 = """

SELECT 
	  DATE_TRUNC(order_date, MONTH)  as Month_, 
	  ROUND(AVG( Total_Keuntungan)) AS Average_daily_spend
FROM (SELECT order_date, SUM(price*quantity) AS Total_Keuntungan
      From  `aqueous-freedom-398208.dkf_okefood.order` O
      JOIN `aqueous-freedom-398208.dkf_okefood.food` F ON O.fk_food_id = F.food_id
      GROUP BY order_date
      )
GROUP BY 1
ORDER BY 1 ASC


  """

#Run the query and write result to a pandas data frame

result = bigquery_client.query(query4)

df = result.to_dataframe()

#View top few rows of result

df.head()
```
